```python
#Desenvolva um programa que receba um número digitado pelo usuário e calcule a soma de
#todos os números de 1 até ao número digitado. Por exemplo, se o usuário digitou o número 4,
#a saída deve ser 10 (1+2+3+4=10)

soma_numeros = 0
numero = int(input("Digite um número:"))
for i in range(1, numero + 1, 1):
    soma_numeros += i
print("A soma é = ", soma_numeros)
```

```python
#Crie um algoritmo que receba um número, conte o número total de dígitos e mostre o resultado.
#Exemplo, se o número é 2021 , então a saída deve ser 4

digitos = int(input("Digite o numeros:"))
contador = 0
while digitos != 0:
    digitos //= 10
    contador+= 1
print("Total de dígitos = ", contador)
```
